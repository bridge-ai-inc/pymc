```python
# data_trans.drop(['nm'], axis=1) #data_trans.columns[0]
# data_trans.columns

data_ref[' %R'] = data_ref[' %R']/100
t = pd.DataFrame(data_trans[' %T']/100)
t.head()
```

```python
data_input = pd.concat([data_ref, t], axis=1, ignore_index=True)
data_input.columns = ['nm','R','T']
data_input.head()
```

```python
data_input.to_csv('merged_data.csv', sep='\t', header=False, index=False)
```

```python
#print(*data_input.values.tolist()[0], sep=" ") #*names, sep=", "
print(' '.join(map(str, data_input.values.tolist()[0])))
```

```python
# create file that has content of headerfile
#loop through merged_data.csv and append to input

copyfile('header.txt', 'input.txt')
d = data_input.values.tolist()
f = open('input.txt', 'a')

for i in range(len(data_input)):
    f.write(' '.join(map(str, d[i])) + '\n')
f.close()
```
